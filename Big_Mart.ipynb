```python
data.replace({'Item_Fat_Content':{'low fat':"Low fat",'LF':"Low Fat",'Low Fat':'Low fat','reg':"Regular"}},inplace=True)
```

LABEL_ENCODING......

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

labe1=LabelEncoder()
```

```python
data['Item_Fat_Content']=labe1.fit_transform(data['Item_Fat_Content'])
```

```python
data['Item_Type']=labe1.fit_transform(data['Item_Type'])
data['Outlet_Identifier']=labe1.fit_transform(data['Outlet_Identifier'])
data['Outlet_Size']=labe1.fit_transform(data['Outlet_Size'])
data['Outlet_Location_Type']=labe1.fit_transform(data['Outlet_Location_Type'])
data['Outlet_Type']=labe1.fit_transform(data['Outlet_Type'])
```

```python
data.head(5)
```

```python
from sklearn.model_selection import train_test_split

x=data.drop(columns=['Item_Outlet_Sales',"Item_Identifier",],axis=1)
```

```python
x.drop(columns=["Item_Visibility"],axis=1,inplace=True)
```

```python
y=data['Item_Outlet_Sales']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(x_test)
```

```python
len(x_train)
```

```python
from xgboost import XGBRFRegressor

xgreg=XGBRFRegressor()
```

```python
xgreg.fit(x_train,y_train)
```

```python
xgreg.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,r2_score

y_pred=xgreg.predict(x_train)
```

```python
len(y_pred)
```

```python
acc=r2_score(y_train,y_pred)
```

```python
f'The Training is : {acc*100}%'
```

```python
Y_pred=xgreg.predict(x_test)
acc_test=r2_score(y_test,Y_pred)
```

```python
f'The Testing is : {acc_test*100}%'
```
